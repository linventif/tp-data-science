```python
import numpy as np

liste_revenus = [1800, 1500, 2200, 3000, 2172, 5000, 1400, 1200, 1100, 1300]
liste_noms = ["Ava", "Bob", "Cao", "Dan", "Eli", "Fay", "Guy", "Hue", "Ian", "Joy"]

# Q2.1. Créez un array nommé revenus à partir de liste_revenus et affichez son contenu. Faites de même pour les noms.
revenus = np.array(liste_revenus)
print(revenus)

# Q2.2. Créez un array hauts_revenus contenant uniquement les revenus supérieurs à 2000C, via une indexation booléenne ou via la fonction where(...) de numpy. Affichez son contenu.
hauts_revenus = revenus > 2000
print(hauts_revenus)

# 2.3. Affichez les noms des personnes gagnant plus de 2000C mensuels, à partir de la fonction where(...).
print(np.array(liste_noms)[np.where(revenus > 2000)])

# Q2.4. Calculez la moyenne des revenus de tous vos clients, via la méthode mean(). Affichez en arrondissant grâce à une f-string.
print(f"La moyenne des revenus est de {revenus.mean()}")

# Q2.5. Calculez et affichez en arrondissant (en KC) la somme des revenus annuels de tous vos clients.
print(f"La somme des revenus est de {revenus.sum() / 12}")

# Q2.6. Le client gagnant 1100C a reçu une augmentation de 150C. De même, Ava a été augmenté de 100C. Ré- percutez ces hausses dans votre structure. Veillez à n"utiliser que les informations à votre disposition pour effectuer chacune des opérations : à partir de la valeur elle-même et l"indexation booléenne d"un côté, à partir du nom à associer avec la fonction where(...) de l"autre.
revenus[liste_noms.index("Ava")] += 100
revenus[np.where(revenus == 1100)] += 150
print(revenus)

# Q2.7. Affichez les noms et revenus de chacun des clients à raison d"un client par ligne, au moyen d"une boucle.
for nom, revenu in zip(liste_noms, liste_revenus):
    print(f"{nom}: {revenu}")
```

```python
# On rappelle que la fonction de densité d’une loi normale d’espérance μ et d’écart-type σ, est notée N (μ, σ), s’exprime
# par :
# f : x 7 → 1
# σ √2 π e− 1
# 2 ( x−μ
# σ )2
# Graphiquement, la probabilité d’un évènement X, notée P [a < X < b], telle que −∞ ≤ a ≤ b ≤ +∞ correspond à
# l’aire de la zone comprise :
# — en abscisses, entre a et b (autrement dit : a < x < b)
# — en ordonnées, entre la l’axe des abscisses et la courbe représentative de la densité (autrement dit : 0 < y < f (x))

import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Q1. Créez une cellule d’import des bibliothèques nécessaires, à savoir numpy et matplotlib.pyplot.
# Already imported above

# Q2. Commençons par visualiser une loi normale.
# Q2.1. Définissez une fonction densité, prenant en paramètre l’espérance μ et l’écart-type σ, qui calcule la valeur
# associée à une variable x prédéfinie.
def densite(mu, sigma, x):
  return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Q2.2. Générez un vecteur x de 1000 valeurs entre 0 et 80, régulièrement espacées, via la fonction linspace(...).
# Ces valeurs serviront à calculer densité(x), et tracer la loi normale correspondante.
x = np.linspace(0, 80, 1000)

# Q2.3. Tracez la courbe de la fonction de la densité de la loi normale, ici N (35, 10), grâce à la fonction plot(...).
# Veillez à ajouter un titre et une légende.
mu, sigma = 35, 10
y = densite(mu, sigma, x)
plt.plot(x, y, label=f'N({mu}, {sigma})')
plt.title('Densité de la loi normale N(35, 10)')
plt.legend()
plt.show()

# Q3. Comprenons maintenant une loi normale.
# Q3.1. Tracez N (35, 10), N (35, 20) et N (35, 5) sur le même graphe. Ajoutez une légende en haut à droite et un titre à
# votre courbe. Constatez l’impact de l’écart-type sur la loi normale.
plt.plot(x, densite(35, 10, x), label='N(35, 10)')
plt.plot(x, densite(35, 20, x), label='N(35, 20)')
plt.plot(x, densite(35, 5, x), label='N(35, 5)')
plt.title('Impact de l\'écart-type sur la loi normale')
plt.legend(loc='upper right')
plt.show()

# Q3.2. De même, tracez N (35, 10), N (45, 10) et N (55, 10) sur le même graphe. Constatez l’impact de l’espérance sur
# la loi normale.
plt.plot(x, densite(35, 10, x), label='N(35, 10)')
plt.plot(x, densite(45, 10, x), label='N(45, 10)')
plt.plot(x, densite(55, 10, x), label='N(55, 10)')
plt.title('Impact de l\'espérance sur la loi normale')
plt.legend(loc='upper right')
plt.show()

# Q3.3. À partir du tracé de la densité, estimez la valeur des probabilités suivantes, où X est une variable aléatoire de
# loi N (35, 10) :
# — P [X < 35]
# — P [X < 45]
# — P [35 < X < 45]
# — P [X > 45]

P_X_less_35 = norm.cdf(35, 35, 10)
P_X_less_45 = norm.cdf(45, 35, 10)
P_35_less_X_less_45 = norm.cdf(45, 35, 10) - norm.cdf(35, 35, 10)
P_X_greater_45 = 1 - norm.cdf(45, 35, 10)

print(f'P[X < 35] = {P_X_less_35}')
print(f'P[X < 45] = {P_X_less_45}')
print(f'P[35 < X < 45] = {P_35_less_X_less_45}')
print(f'P[X > 45] = {P_X_greater_45}')

# Q4. Nous allons maintenant simuler et représenter des échantillons pour la loi normale N (35, 10) pour comprendre
# l’impact des différents paramètres.
# Q4.1. Générez un vecteur nommé ages100 contenant 100 éléments, qui vérifie cette loi statistique. Il convient
# d’invoquer la distribution .normal(...) sur un générateur aléatoire.
ages100 = np.random.normal(35, 10, 100)

# Q4.2. Tracez un histogramme (via hist(..., density=True, alpha=0.25) représentant l’échantillon. Vérifiez
# sa forme en superposant le graphe de la densité de la loi qu’il est censé suivre (toujours sur 1000 points entre 0 et 80).
plt.hist(ages100, bins=30, density=True, alpha=0.25, label='Histogramme ages100')
plt.plot(x, densite(35, 10, x), label='N(35, 10)')
plt.title('Histogramme de ages100 et densité de N(35, 10)')
plt.legend()
plt.show()

# Q4.3. Afin de constater l’impact de la taille de l’échantillon, générez des vecteurs nommés ages10, ages10K et
# ages1M contenant respectivement 10, 100000 et 1e6 éléments, qui vérifient cette loi statistique.
ages10 = np.random.normal(35, 10, 10)
ages10K = np.random.normal(35, 10, 10000)
ages1M = np.random.normal(35, 10, 1_000_000)

# Q4.4. Tracez les différents histogrammes sur le même graphe ainsi que la loi de densité pour observer l’impact de la
# taille de l’échantillon.
plt.hist(ages10, bins=30, density=True, alpha=0.25, label='ages10')
plt.hist(ages10K, bins=30, density=True, alpha=0.25, label='ages10K')
plt.hist(ages1M, bins=30, density=True, alpha=0.25, label='ages1M')
plt.plot(x, densite(35, 10, x), label='N(35, 10)')
plt.title('Impact de la taille de l\'échantillon sur l\'histogramme')
plt.legend()
plt.show()

# Q4.5. Reste à évaluer l’importance de finesse de l’histogramme : le paramètre bins. Pour l’échantillon ages10K,
# tracez les histogrammes pour la loi normale N (35, 10) obtenus pour des valeurs de bins de 10, 100 et 1000. Constatez
# l’effet ainsi obtenu.
plt.hist(ages10K, bins=10, density=True, alpha=0.25, label='bins=10')
plt.hist(ages10K, bins=100, density=True, alpha=0.25, label='bins=100')
plt.hist(ages10K, bins=1000, density=True, alpha=0.25, label='bins=1000')
plt.plot(x, densite(35, 10, x), label='N(35, 10)')
plt.title('Impact du paramètre bins sur l\'histogramme de ages10K')
plt.legend()
plt.show()
```
